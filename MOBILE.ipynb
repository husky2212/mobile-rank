```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
URL1 = 'https://www.gsmarena.com/xiaomi_12_lite-11472.php';
#for url in range(0, 5):
page = requests.get(URL1)
soup = BeautifulSoup(page.text, 'html.parser')
titles = soup.find_all(attrs={'class': 'specs-phone-name-title', 'data-spec': 'modelname'})
Titles = titles[0].text.replace('h1'," ")
launch = soup.findAll(attrs = {'class': 'nfo', 'data-spec': 'year'})
Launch = launch[0].text.replace('td'," ")
resolution = soup.findAll(attrs = {'class': 'nfo', 'data-spec': 'displayresolution'})
Resolution = resolution[0].text.replace('td'," ")
os = soup.findAll(attrs = {'class': 'nfo', 'data-spec': 'os'})
Os = os[0].text.replace('td'," ")
Os.replace(',' , " ")
chipset = soup.findAll(attrs = {'class': 'nfo', 'data-spec': 'chipset'})
Chipset = chipset[0].text.replace('td'," ")
cpu = soup.findAll(attrs = {'class': 'nfo', 'data-spec': 'cpu'})
Cpu = cpu[0].text.replace('td'," ")
memory = soup.findAll(attrs = {'class': 'nfo', 'data-spec':'internalmemory'})
Memory = memory[0].text.replace('td'," ")
maincamera = soup.findAll(attrs = {'class': 'nfo', 'data-spec':'cam1modules'})
Maincamera = maincamera[0].text.replace('td\r\n'," ")
frontcamera = soup.findAll(attrs = {'class': 'nfo', 'data-spec':'cam2modules'})
Frontcamera = frontcamera[0].text.replace('td'," ")
wlan = soup.findAll(attrs = {'class': 'nfo', 'data-spec':'wlan'})
WLAN = wlan[0].text.replace('td', " ")
bluetooth = soup.findAll(attrs = {'class': 'nfo', 'data-spec':'bluetooth'})
Bluetooth = bluetooth[0].text.replace('td'," ")
```

```python
data1 = {
    "titles": [Titles],
    "Launch":[Launch],
    "Resolution":[Resolution],
    "Os":[Os],
    "Chipset":[Chipset],
    "Cpu":[Cpu],
    "Memory":[Memory],
    "Maincamera":[Maincamera],
    "Frontcamera":[Frontcamera],
    "WLAN":[WLAN],
    "Bluetooth":[Bluetooth]
    
}

data1 = pd.DataFrame(data1)
data1
```
